```python
import os

# Specify the main directory containing training, testing, and validation directories
base_directory = 'D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces'

# Specify subdirectories for training, testing, and validation
training_directory = os.path.join(base_directory, 'training')
testing_directory = os.path.join(base_directory, 'testing')
validation_directory = os.path.join(base_directory, 'validation')

# Function to delete files within a directory
def delete_files(directory):
    for class_folder in os.listdir(directory):
        class_folder_path = os.path.join(directory, class_folder)

        # Check if it is a directory
        if os.path.isdir(class_folder_path):
            # Iterate through files in the class folder and delete them
            for filename in os.listdir(class_folder_path):
                file_path = os.path.join(class_folder_path, filename)
                try:
                    if os.path.isfile(file_path):
                        os.remove(file_path)
                        print(f"Removed file: {file_path}")
                except Exception as e:
                    print(f"Error deleting file {file_path}: {e}")

# Delete files in the training directory
delete_files(training_directory)

# Delete files in the testing directory
delete_files(testing_directory)

# Delete files in the validation directory
delete_files(validation_directory)

print("Files deleted successfully.")
```

```python
import os
import shutil

# Specify the source directory containing all testing images
training_directory = 'D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Training'

# Specify the destination directory for training images
testing_directory = 'D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Testing'

# Iterate through each class folder in the testing directory
for class_folder_test in os.listdir(training_directory):
    class_folder_path_test = os.path.join(training_directory, class_folder_test)
    class_name=os.path.basename(class_folder_path_test)
    files=os.listdir(class_folder_path_test)
    for i in range(1):
        image_name = files[i]
#         print(image_name)
        source_path = os.path.join(training_directory,class_name, image_name)
        destination_path = os.path.join(testing_directory,class_name, image_name)
        shutil.move(source_path, destination_path)
        print(f"Moved file: {source_path} to {destination_path}")

print("Images moved back from testing to training successfully.")
```
