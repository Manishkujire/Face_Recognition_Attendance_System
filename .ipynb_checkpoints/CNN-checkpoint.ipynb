```python
import os
import cv2
import numpy as np

import glob
def get_image_data(directory_path):
    label = []
    image_paths = []

    for subdir in glob.glob(os.path.join(directory_path, '*')):
        for subdirdir in glob.glob(os.path.join(subdir, '*.jpg')):
            img_arr = cv2.imread(subdirdir, 1)
            img_gray = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)
            
            filename = os.path.basename(subdirdir)
            label.append(filename.split(".")[0].split("_")[1])
            image_paths.append(img_gray)

    return  np.array(image_paths),np.array(label)

```

```python
# Example usage:
X_train, y_train = get_image_data('D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Training')
X_test, y_test = get_image_data('D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Testing')

# Print the shape of the resulting arrays
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)


# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models

# Assuming your image dimensions are 80x70 pixels
img_width, img_height = 80, 70

# Assuming you have 1 channel (grayscale)
input_shape = (img_width, img_height, 1)  # Update input shape here

# Define the CNN model
model = models.Sequential()

# Reshape layer for input
model.add(layers.InputLayer(input_shape=input_shape))

# Convolutional layer 1
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional layer 2
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional layer 3
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten layer
model.add(layers.Flatten())

# Fully connected layer 1
model.add(layers.Dense(256, activation='relu'))

# Fully connected layer 2 (output layer)
model.add(layers.Dense(1, activation='sigmoid'))  # Adjust the number of output neurons based on your classes

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])


# Display the model summary
model.summary()

# Assuming you have training and testing data (X_train, y_train, X_test, y_test)
# Adjust batch size and epochs based on your data and computational resources
batch_size = 4
epochs = 10

# Train the model
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    rescale=1./255  # Normalize pixel values to [0, 1]
)

# Load and augment training data
train_datagen = datagen.flow_from_directory(
    'D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Training',
    target_size=(80, 70),
    color_mode='grayscale',
    class_mode='binary',  # Adjust based on your classification task
    batch_size=32,
    shuffle=True
)

# Load validation data (assuming you have a separate validation set)
val_datagen = datagen.flow_from_directory(
    'D:\\Downloads\\Face-Recognition-master (1)\\Face-Recognition-master\\Datasets\\Faces\\Testing',
    target_size=(80, 70),
    color_mode='grayscale',
    class_mode='binary',
    batch_size=32,
    shuffle=False  # No need to shuffle the validation set
)

# Define the CNN model
model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(80, 70, 1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))  # Dropout for regularization
model.add(layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display the model summary
model.summary()

# Train the model with augmented data
history = model.fit(
    train_datagen,
    epochs=20,
    validation_data=val_datagen
)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(val_datagen)
print(f'Test accuracy: {test_acc}')
```
